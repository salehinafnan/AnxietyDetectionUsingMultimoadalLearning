```python
dass_anxiety_data = pd.DataFrame()
dass_anxiety_data["Dryness_of_mouth"] = dass.Dryness_of_mouth
dass_anxiety_data["Breathing_difficulty"] = dass.Breathing_difficulty
dass_anxiety_data["Experienced_trembling"] = dass.Experienced_trembling
dass_anxiety_data["Worried_about_panic_and_make_a_fool_of_myself"] = dass.Worried_about_panic_and_make_a_fool_of_myself
dass_anxiety_data["Close_to_panic"] = dass.Close_to_panic
dass_anxiety_data["Aware_of_the_action_of_heart_in_the_absence_of_physical_exertion"] = dass.Aware_of_the_action_of_heart_in_the_absence_of_physical_exertion
dass_anxiety_data["Scared_without_any_good_reason"] = dass.Scared_without_any_good_reason
```

```python
def scores(dass_without_duplicates):
    col = list(dass_without_duplicates)
    dass_without_duplicates["Scores"] = dass_without_duplicates[col].sum(axis=1)
    return dass_without_duplicates
```

```python
score_data = scores(dass_without_duplicates)
score_data
```

```python
score_data.head(20)
```

```python
dass_without_duplicates["d_score"] = dass_without_duplicates['Scores']*2
```

```python
dass_without_duplicates.head(20)
```

```python
def condition(x):
    if x<=7:
        return 0 # 'Normal'
    if  8<=x<=9:
        return 1 # 'Mild'
    if 10<=x<=14:
        return 2 # 'Moderate'
    if 15<=x<=19:
        return 3 # 'Severe'
    if x>=20:
        return 4 # 'Extremely Severe'

dass_without_duplicates['condition']=dass_without_duplicates['d_score'].apply(condition)
dass_without_duplicates
```

```python
dass_without_duplicates.isnull().sum()
```

```python
dass_without_duplicates = dass_without_duplicates.drop('Scores',axis=1)
dass_without_duplicates = dass_without_duplicates.drop('d_score',axis=1)
```

```python
x = dass_without_duplicates.drop('condition',axis=1)
y = dass_without_duplicates['condition']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=116,random_state=5,shuffle=True)
```

```python
import sklearn
from imblearn.over_sampling import RandomOverSampler

xtrain_ros, ytrain_ros = RandomOverSampler(random_state=2).fit_resample(xtrain, ytrain)

print('After OverSampling, the shape of train_x: {}'.format(xtrain_ros.shape))
print('After OverSampling, the shape of train_y: {}'.format(ytrain_ros.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

models = []
models.append(['XGBClassifier',XGBClassifier()])
models.append(['KNeigbors',KNeighborsClassifier()])
models.append(['GaussianNB',GaussianNB()])
models.append(['SVM',SVC()])
models.append(['DecisionTree',DecisionTreeClassifier(random_state=0)])
models.append(['RandomForest',RandomForestClassifier(random_state=0)])
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score,ConfusionMatrixDisplay,precision_score,recall_score,f1_score,classification_report,roc_curve,auc,precision_recall_curve,average_precision_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
lst_1 = []
for m in range(len(models)):
    lst_2 = []
    per_model = models[m][1]
    per_model.fit(xtrain_ros,ytrain_ros)
    y_pred = per_model.predict(xtest)
    cr = classification_report(ytest,y_pred)
    cf = confusion_matrix(ytest,y_pred)
    cf_df = pd.DataFrame(cf,
                    index=[0,1,2,3,4],
                    columns=[0,1,2,3,4])
    sns.heatmap(cf,annot=True)
    plt.title(models[m][0])
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.show()
    
    precision = precision_score(ytest,y_pred,average='weighted',zero_division=0)
    recall = recall_score(ytest,y_pred,average='weighted',zero_division=0)
    f1 = f1_score(ytest,y_pred,average='weighted',zero_division=0)
    
    print(str(per_model)+" : \n")
    print("")
    print(cr)
    print("-"*25)
    print("")
    lst_2.append(models[m][0])
    lst_2.append(accuracy_score(ytest,y_pred)*100)
    lst_2.append(100-accuracy_score(ytest,y_pred)*100)
    lst_2.append(precision)
    lst_2.append(recall)
    lst_2.append(f1)
    lst_1.append(lst_2)
```

```python
df = pd.DataFrame(lst_1,columns=['Model','Accuracy','Error Rate','Precision','Recall','F1 Score'])

df.sort_values(by=['Accuracy'],inplace=True,ascending=True)
df
```

```python
grid_models = [
    (KNeighborsClassifier(), [{'n_neighbors': [5, 7, 8, 10], 'metric': ['euclidean', 'manhattan', 'chebyshev', 'minkowski']}]),
    (DecisionTreeClassifier(), [{'criterion': ['gini', 'entropy'], 'random_state': [0]}]),
    (RandomForestClassifier(), [{'n_estimators': [100, 150, 200], 'criterion': ['gini', 'entropy'], 'random_state': [0]}]),
    (XGBClassifier(), [{'n_estimators': [100, 150, 200], 'learning_rate': [0.1, 0.01, 0.001], 'max_depth': [3, 5, 7]}]),
    (SVC(), [{'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf', 'poly'], 'gamma': ['scale', 'auto']}]),
    (GaussianNB(), [{}])  # No hyperparameters to tune for Gaussian Naive Bayes
]
```

```python
for i,j in grid_models:
    grid = GridSearchCV(estimator=i,param_grid = j, scoring = 'accuracy',cv = 10)
    grid.fit(xtrain_ros,ytrain_ros)
    best_accuracy = grid.best_score_
    best_param = grid.best_params_
    print(' {}: \n Best Accuracy: {:.0f} %'.format(i,best_accuracy*100))
    print(' Best Param: {}'.format(best_param))
    print('')
    print('-'*25)
    print('')
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming grid_models contains a list of model instances (i) and corresponding parameter grids (j)
best_accuracy = 0
best_params = {}
best_model = None

for i, j in grid_models:
    grid = GridSearchCV(estimator=i, param_grid=j, scoring='accuracy', cv=10)
    grid.fit(xtrain_ros, ytrain_ros)
    accuracy = grid.best_score_
    params = grid.best_params_
    print('{}: \n Best Accuracy: {:.0f} %'.format(i.__class__.__name__, accuracy * 100))
    print('Best Params: {}'.format(params))
    print('')
    print('-' * 25)
    print('')

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_params = params
        best_model = i  # Store the best model instance, not the best_estimator_

# Create the best model with the best hyperparameters
best_model = best_model.set_params(**best_params)

# Train the best model on the training data
best_model.fit(xtrain_ros, ytrain_ros)

# Make predictions on the test data
y_pred = best_model.predict(xtest)

# Calculate evaluation metrics
acc = accuracy_score(ytest, y_pred)
precision = precision_score(ytest, y_pred, average='weighted', zero_division=0)
recall = recall_score(ytest, y_pred, average='weighted', zero_division=0)
f1 = f1_score(ytest, y_pred, average='weighted', zero_division=0)

# Display confusion matrix
cf = confusion_matrix(ytest, y_pred)
cf_df = pd.DataFrame(cf, index=[0, 1, 2, 3, 4], columns=[0, 1, 2, 3, 4])
sns.heatmap(cf, annot=True)
plt.title('Confusion Matrix')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()

# Print evaluation metrics
print("Best Accuracy:", best_accuracy)
print("Accuracy:", acc)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
